```python
import numpy as np
import matplotlib.pyplot as plt

# Number of random numbers to generate
N = 10000

# Generate uniform random numbers
u1 = np.random.rand(N)
u2 = np.random.rand(N)

# Box-Muller transform
z0 = np.sqrt(-2*np.log(u1)) * np.cos(2*np.pi*u2)
z1 = np.sqrt(-2*np.log(u1)) * np.sin(2*np.pi*u2)

# Concatenate z0 and z1 to get 10000 samples from the Gaussian distribution
z = np.concatenate((z0, z1))

# Create a density histogram of the generated samples
plt.hist(z, bins=50, density=True, alpha=0.6, color='b', edgecolor='black')

# Plot the Gaussian PDF for comparison
x = np.linspace(-4, 4, 1000)
pdf = 1/np.sqrt(2*np.pi) * np.exp(-0.5*x**2)
plt.plot(x, pdf, 'r', linewidth=2)

plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Histogram of Gaussian Distribution vs Gaussian PDF')
plt.legend(['Gaussian PDF', 'Histogram'])

plt.show()
```
