```python
import numpy as np
import matplotlib.pyplot as plt
import time

def lcg(seed, a, c, m, size):
    random_numbers = []
    X = seed
    for _ in range(size):
        X = (a * X + c) % m
        random_numbers.append(X / m)  # Normalize to [0, 1)
    return random_numbers

# Parameters for the LCG
seed = 41
a = 1664525
c = 1013904223
m = 2**32
size = 10000

# Measure the time taken to generate random numbers
start_time = time.time()
random_numbers = lcg(seed, a, c, m, size)
end_time = time.time()
elapsed_time = end_time - start_time
print(f"Time taken to generate {size} random numbers: {elapsed_time} seconds")

# Plotting the histogram
plt.figure(figsize=(10, 6))
plt.hist(random_numbers, bins=50, density=True, alpha=0.6, color='g', label='LCG Random Numbers')

# Plotting the uniform PDF
x = np.linspace(0, 1, 100)
uniform_pdf = np.ones_like(x)
plt.plot(x, uniform_pdf, 'r-', lw=2, label='Uniform PDF')

# Adding labels and legend
plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Density Histogram of LCG Random Numbers vs Uniform PDF')
plt.legend()

# Show the plot
plt.show()
```
