```python
import numpy as np
import matplotlib.pyplot as plt

# Read the data from the text file
data = np.loadtxt('exp_deviates.txt')

# Plot the density histogram of the data
plt.hist(data, bins=50, density=True, alpha=0.6, color='g', label='Histogram of generated data')

# Plot the theoretical exponential PDF
x = np.linspace(0, max(data), 1000)
mean = 0.5
lam = 1 / mean
pdf = lam * np.exp(-lam * x)
plt.plot(x, pdf, 'r-', label='Exponential PDF with mean 0.5')

# Add labels and legend
plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Density Histogram and Exponential PDF')
plt.legend()

# Show the plot
plt.show()
```
