```python
import numpy as np
import matplotlib.pyplot as plt

# True solution
def true_solution(t):
    return np.exp(1 - 9 * t)

# Explicit Euler method
def explicit_euler(h, t_final):
    num_steps = int(t_final / h)
    t_values = np.linspace(0, t_final, num_steps + 1)
    y_values = np.zeros(num_steps + 1)
    y_values[0] = np.exp(1)  # Initial condition

    for i in range(num_steps):
        y_values[i + 1] = y_values[i] + h * (-9 * y_values[i])

    return t_values, y_values

# Implicit Euler method
def implicit_euler(h, t_final):
    num_steps = int(t_final / h)
    t_values = np.linspace(0, t_final, num_steps + 1)
    y_values = np.zeros(num_steps + 1)
    y_values[0] = np.exp(1)  # Initial condition

    for i in range(num_steps):
        y_values[i + 1] = y_values[i] / (1 + 9 * h)

    return t_values, y_values

# Parameters
h = 0.24
t_final = 3

# Compute solutions
t_values, y_explicit = explicit_euler(h, t_final)
t_values, y_implicit = implicit_euler(h, t_final)
y_true = true_solution(t_values)

# Plot results
t1_values = np.linspace(0, t_final, 100)  # Generate 100 points between 0 and 1
y_values = true_solution(t1_values) # Compute y values for each t

plt.figure(figsize=(12, 6))
plt.plot(t1_values, y_values, label='True Solution', color='black')
plt.plot(t_values, y_explicit, label='Explicit Euler', linestyle='--', marker='o', color='blue')
plt.plot(t_values, y_implicit, label='Implicit Euler', linestyle='--', marker='x', color='red')
plt.xlabel('t')
plt.ylabel('y(t)')
plt.legend()
plt.grid(True)
plt.title('Comparison of Explicit and Implicit Euler Methods')
plt.show()
```
