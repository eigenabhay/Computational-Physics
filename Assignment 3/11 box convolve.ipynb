```python
import numpy as np
import matplotlib.pyplot as plt

# Define the box function
def box_function(x):
    return np.where(np.logical_and(x > -1, x < 1), 1, 0)

# Create x values
x = np.linspace(-5, 5, 512)
dx = x[1]-x[0]
# Compute the convolution
convolution = np.convolve(box_function(x), box_function(x), mode='same')*dx

# Plot the box function and its convolution
plt.figure(figsize=(10, 6))
plt.plot(x, box_function(x), label='Box Function')
plt.plot(x, convolution, label='Convolution')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Convolution of Box Function with Itself')
plt.legend()
plt.grid(True)
plt.show()
```
